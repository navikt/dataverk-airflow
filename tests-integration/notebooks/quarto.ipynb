## matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#| label: fig
#| fig-cap: Dette er en figur, og denne figurteksten er oppdatert.
#| include: true
r = np.arange(0, 2, 0.01)
theta = 2 * np.pi * r
fig, ax = plt.subplots(
  subplot_kw = {'projection': 'polar'} 
)
ax.plot(theta, r)
ax.set_rticks([0.5, 1, 1.5, 2])
ax.grid(True)
plt.show()
```
